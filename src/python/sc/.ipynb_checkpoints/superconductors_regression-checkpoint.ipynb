```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
plt.style.use('ggplot')
```

```python
data = pd.read_csv("SC_short_Matminer_features.csv")
data.shape
```

```python
data.head()
```

```python
X = data[data.columns[5:]]
X["Cuprate"] = data["Cuprate"]
X["Fe-based"] = data["Fe-based"]
y = data["Tc"] #y = result["Tc"]

print(X.shape)
print(y.shape)
X.head()
```

```python
X_high=X[y>10]
y_high=y[y>10]
print("Number of compounds with T_c> K: ", X_high.shape[0])
```

```python
# Create train and test datasets in ratio 0.7/0.3

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_high, y_high,  test_size=0.3,  random_state=11)
```

```python
from sklearn.ensemble import RandomForestRegressor

preds = X.columns[0:132]
print(X_test[preds].shape)

rf = RandomForestRegressor(n_estimators=100, random_state =51, oob_score = True)

rf.fit( X_train[preds], y_train) #[index_30]
print(rf.oob_score_ , rf.score(X_test[preds], y_test))
# X_test_pred = rf.predict(X_test)
```

```python
rf
```

```python
# Optimize the model
# Use GridSearch to optimize the hyper-parameters. Important parameters for random forest models:

# number of trees, depth of each tree, how many features to consider at each split, min number of instances in a leaf

#Slow!!! Do not have to run.
```

```python
#from sklearn.grid_search import GridSearchCV
from sklearn.datasets import load_digits
from sklearn.model_selection import GridSearchCV


model = RandomForestRegressor(random_state=30)
param_grid = {"n_estimators"      : [500],
           "max_features"      : [0.2, 0.3, 0.4], #,
              'max_depth': [30, 35, None],
            'min_samples_split':range(2, 6, 2)
         # "bootstrap": [True, False]
             }
grid_search = GridSearchCV(model, param_grid, n_jobs=-1, cv=5)
grid_search.fit(X_train[preds], y_train)

print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
# plot the results

rf = RandomForestRegressor(n_estimators=500, random_state =30, oob_score = True, 
                           max_depth=30, max_features= 0.2, min_samples_split=2)

rf.fit( X_train[preds], y_train) #[index_30]
y_test_pred = rf.predict(X_test[preds])
print(rf.oob_score_ , rf.score(X_test[preds], y_test))
```
